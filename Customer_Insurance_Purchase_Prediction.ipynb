```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Social_Network_Ads.csv')
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

__Logistic Regression__

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_test=lr.predict(x_test)
y_pred_train=lr.predict(x_train)
```

```python
print(np.concatenate((y_test.reshape(len(y_test),1),y_pred_test.reshape(len(y_pred_test),1)),1))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

cm = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix:\n", cm)

accuracy = accuracy_score(y_test, y_pred_test)
print("Accuracy:", accuracy)

precision = precision_score(y_test, y_pred_test)
print("Precision:", precision)

recall = recall_score(y_test, y_pred_test)
print("Recall:", recall)

f1 = f1_score(y_test, y_pred_test)
print("F1-Score:", f1)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from matplotlib.colors import ListedColormap


# Inverse transform for plotting
x_set = sc.inverse_transform(x_train)
y_set = y_train

# Mesh grid for decision boundary
x1_vals = np.linspace(x_set[:, 0].min() - 2, x_set[:, 0].max() + 2, num=100)
x2_vals = np.linspace(x_set[:, 1].min() - 200, x_set[:, 1].max() + 200, num=100)
x1, x2 = np.meshgrid(x1_vals, x2_vals)

# Predict on scaled grid points
grid_points = np.array([x1.ravel(), x2.ravel()]).T
Z = lr.predict(sc.transform(grid_points))  # Scale grid points
Z = Z.reshape(x1.shape)

# Plot
plt.contourf(x1, x2, Z, alpha=0.75, cmap=ListedColormap(('olive', '#FF9999')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1], 
                c=ListedColormap(('olive', '#FF9999'))(i), label=j)

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
